```python
import math
import os
import glob
import json
from PIL import Image
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
%matplotlib inline
```

```python
def parse_json(json_file):
    polygon_dict = {}
    for i in json_file['data_set_info']['data']:
        label = i['value']['object_Label']['attribute']
        polygon_list = []
        for j in i['value']['points']:
            polygon_list.append( [j['x'],j['y']])
            polygon_dict[label] = polygon_list
            
    polygon_list = np.array(polygon_list, np.int32)
    return polygon_dict
```

```python
def draw_label(polgon_dict,filename):
    bg = np.zeros((1080, 1920, 3), np.uint8)
    try:
        points1 = np.array(polgon_dict['violation'], np.int32)
    except KeyError:
        points1 = np.array(polgon_dict['danger'], np.int32)
    try:
        points2 = np.array(polgon_dict['single_solid'],np.int32)
    except KeyError:
        points2 = np.array(polgon_dict['double_solid'],np.int32)
    img = cv2.fillPoly(bg, [points2], (0,0,255))
    img = cv2.fillPoly(bg, [points1], (0, 255, 0))
    img = Image.fromarray(img)
    img.save('.' + f'{filename}.png')
```

```python
label_path = './20211110_DETECT_7655/ANNOTATION/'
json_files = glob.glob(label_path + '*/*/*.json')
```

```python
len(json_files)
```

```python
for i in range(len(json_files)):
    json_path = json_files[i]
    with open(json_path, encoding="utf8") as f:
        print(json_path.split('.')[-2])
        json_file = json.load(f)
        polgon_dict = parse_json(json_file)
        draw_label(polgon_dict,json_path.split('.')[-2])
```
